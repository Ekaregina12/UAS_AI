```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

```python
# Membaca dataset Iris
dataset = pd.read_csv('iris.csv')
```

```python
# Mengambil kolom target (label)
y = dataset.iloc[:, -1].values

# Mengubah label kelas menjadi angka menggunakan LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Menampilkan label yang diubah menjadi angka
print(y_encoded)
```

```python
# Memisahkan fitur dan label
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# Membagi dataset menjadi data pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
print(X_train.shape)

# Memastikan jumlah kolom adalah 3
if X_train.shape[1] == 3:
    print("Jumlah kolom adalah 3")
else:
    print("Jumlah kolom bukan 3")
```

```python
# Menghilangkan fitur yang tidak diperlukan (misalnya, fitur ke-4)
X_train_processed = X_train[:, :3]
X_test_processed = X_test[:, :3]

print("Dimensi X_train_processed:", X_train_processed.shape)
print("Dimensi X_test_processed:", X_test_processed.shape)
```

```python
# Memeriksa dimensi X_train
print("Dimensi X_train sebelum:", X_train.shape)

# Memeriksa dimensi y_train
print("Dimensi y_train sebelum:", y_train.shape)

# Mengubah dimensi y_train menjadi (jumlah_sampel,)
y_train = y_train.reshape(-1)

# Memeriksa kembali dimensi y_train setelah diubah
print("Dimensi y_train setelah:", y_train.shape)
```

```python
# Membangun model ANN
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(3,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
# Mengompilasi model dengan optimizer Adam dan loss function 'sparse_categorical_crossentropy'
optimizer = optimizers.Adam()
model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Melatih model dengan data pelatihan
model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1)
```
